```python
from src.entropy.entropy import *
from src.entropy.entropy_helpers import *
from src.entropy.plots import plot_2d_partitions, plot_3d_partitions
import numpy as np
```

```python
np.random.seed(0)
N = 200
X = np.random.uniform(0, 1, N)
Y = X + np.random.normal(0, 0.1, N)
Z = Y + np.random.normal(0, 0.1, N)

# Rank the data (ordinal transformation)
X_ranks = X.argsort().argsort() + 1
Y_ranks = Y.argsort().argsort() + 1
Z_ranks = Z.argsort().argsort() + 1
```

```python
data2D = np.array([X_ranks, Y_ranks]).T
rank_max = max(X_ranks.max(), Y_ranks.max())
partitions = dv_partition_nd(data2D, mins=np.array([1, 1]), maxs=np.array([rank_max, rank_max]))
plot_2d_partitions(partitions, X_ranks, Y_ranks)
```

```python
data3D = np.array([X_ranks, Y_ranks, Z_ranks]).T
rank_max = max(X_ranks.max(), Y_ranks.max(), Z_ranks.max())
partitions = dv_partition_nd(data3D, mins=np.array([1, 1, 1]), maxs=np.array([rank_max, rank_max, rank_max]))
plot_3d_partitions(partitions, X_ranks, Y_ranks, Z_ranks)
```
