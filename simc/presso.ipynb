```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
start = 2000
end = 3500
abp_rest = pd.read_csv('..\data\CONTROL_BREATHING_RECORDINGS\OCHOTNIK_1\CLEAN_DATA_BASELINE\OCH_1_CLEAN_CB_BASELINE.csv', sep=';', decimal=',')['abp_cnap[mmHg]'].values
abp_6b = pd.read_csv('..\data\CONTROL_BREATHING_RECORDINGS\OCHOTNIK_1\CLEAN_DATA_6_BREATHS\OCH_1_CLEAN_CB_6.csv', sep=';', decimal=',')['abp_cnap[mmHg]'].values
abp_10b = pd.read_csv('..\data\CONTROL_BREATHING_RECORDINGS\OCHOTNIK_1\CLEAN_DATA_10_BREATHS\OCH_1_CLEAN_CB_10.csv', sep=';', decimal=',')['abp_cnap[mmHg]'].values
abp_15b = pd.read_csv('..\data\CONTROL_BREATHING_RECORDINGS\OCHOTNIK_1\CLEAN_DATA_15_BREATHS\OCH_1_CLEAN_CB_15.csv', sep=';', decimal=',')['abp_cnap[mmHg]'].values
```

```python
print(np.mean(abp_6b))
print(np.mean(abp_10b))
print(np.mean(abp_15b))
```

```python
plt.figure(figsize=(15, 6), dpi=300)
plt.plot(abp_6b[start:end], label='6', color='blue', alpha=0.5)
plt.plot(abp_10b[start:end], label='10', color='#E1780A', alpha=0.5)
plt.plot(abp_15b[start:end], label='15', color='green', alpha=0.5)
plt.legend(title='Breaths per minute')
plt.title('ABP signal during controlled breathing', fontsize=16)
plt.xlabel('Samples', fontsize=14)
plt.ylabel('ABP [mmHg]', fontsize=14)
plt.show()
```
